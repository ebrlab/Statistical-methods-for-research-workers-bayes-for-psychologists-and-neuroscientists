# Bayesian estimation equivalent of the one-way ANOVA

# Bayesian inference
<font size = "3"> Following the quick description of the classic one-way ANOVA above its important to keep in mind that Bayesian inference is derived from the application of Bayes rule $P(\theta \mid y) = \large \frac{P(y \mid \theta) \, P(\theta)}{P(y)}$ and as such while the following description of the Bayesian model is an equivalent to the one-way ANOVA, it is fundamentally different, because it uses fully probabilistic modelling and the inferences are not based on sampling distributions.</font>
    
<font size = "1"> For a fuller description see the Practicing Bayesian statistics markdown file within the Github repository.</font>

# Steps of Bayesian data analysis

<font size = "3"> Kruscke (2015) offers a step by step formulation for how to conduct a Bayesian analysis:

1. Identify the relevant data for the question under investigation.

2. Define the descriptive (mathematical) model for the data.

3. Specify the Priors for the model. If a scientific research publication is the goal, the priors must be accepted by a skeptical audience. This can be achieved by using prior predcitve checks to ascertain if the priors are reasonable.

4. Using Bayes rule, estimate the posterior for the model parameters using the likelihood and priors. Then use the posterior to conduct your inference.

5. Conduct model checks. i.e. Posterior predcitive checks.</font> 

<font size = "1">This notebook will follow this approach generally.</font>

4. Reactivation only: Participants only watch the trauma film

# Step 4 - Use Bayes rule

# Stan model of Bayesian One-way ANOva

```python
One_ANOVA = """
data{

int<lower = 0> N; // Number od data points
vector[N] y; // Dependent varaible data
int K; // number of groups
 // array of interger group indicator variables
int<lower = 1, upper = K> x[N];

}

parameters{

// By specifying the lower bound the normal for the priors
// become half normal prior distributions
real<lower = 0> sigma; // homoscedastic standard deviation
vector<lower = 0>[K] mu; // mean parameter for each group
}

model{

//priors 
mu ~ normal(0, 100); // prior on each group
sigma ~ normal(0,1);

//Likelihood
y ~ normal(mu[x], sigma);

}

generated quantities{

// Calculate the pairwise comparisons
real group1vsgroup2 = mu[1] - mu[2];
real group1vsgroup3 = mu[1] - mu[3];
real group1vsgroup4 = mu[1] - mu[4]; 
real group2vsgroup3 = mu[2] - mu[3]; 
real group2vsgroup4 = mu[2] - mu[4];
real group3vsgroup4 = mu[3] - mu[4];

// PPC
vector[N] yrep; 

for (i in 1:N) {
yrep[i] = normal_rng(mu[x[i]], sigma);
 }
}
 """
```

```python
sm = ps.StanModel(model_code=One_ANOVA)
```

```python
data = {"N": len(df),
        "y": df["Day_Zero_Number_of_Intrusions"].values,
        "K":  max(df["Condition"].values,
        "x": df["Condition"].values
       }
```

```python
fit = sm.sampling(data = data, iter = 2000, chains=4, seed= 302675)
```

```python
summary = fit.summary()
fit_df = pd.DataFrame(summary['summary'], 
                  columns = summary['summary_colnames'], 
                  index = summary['summary_rownames'])
```

# Post model fit-visualisations - Bayesian one sample Z-test

## Posterior distributions plots

```python
# Using arviz built in Bayesian anlysis plot_posterior funcion to display the posterior probabilty distributions
# for the fitted model parameter estimates.
az.plot_posterior(fit, var_names=("mu", "sigma"));
```

## Autocorrelation plots

```python
# Using the arviz package the autocorrelation of the 4 MCMC chains can be plotted.
az.plot_autocorr(fit, var_names=("mu", 'sigma'));
```

## MCMC traceplots

```python
az.plot_trace(fit,var_names=("mu", "sigma"));
```

The MCMC traceplots show good mixing of the chains, showing a "fuzzy caterpillar".

# Step 5 - Posterior predictive checks

```python
#Convert pystan fit object to IO for Arviz ppc functions.
postFit = az.from_pystan(
                posterior= fit,
                posterior_predictive='yrep',
                observed_data=["y"])
```

```python
# Plot posterior simulated data sets for posterior predictive check
az.plot_ppc(postFit, data_pairs = {"y" : "yrep"}, num_pp_samples= 100);
```

PPC shows a poor model fit with the data and that the data is likely not normally distributed. see caveat section for fuller explanation
