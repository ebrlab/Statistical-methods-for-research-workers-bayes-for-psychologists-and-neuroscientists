# Bayesian estimation equivalent of multiple regression

## The classic Multiple regression

# Bayesian inference
<font size = "3"> Following the quick description of the classic multiple regression above its important to keep in mind that Bayesian analysis inference are all derived from the application of Bayes rule $P(\theta \mid y) = \large \frac{P(y \mid \theta) \, P(\theta)}{P(y)}$ and as such while the following description of the Bayesian model is an equivalent to the classic multiple regression, it is fundamentally different, because its uses fully probabilistic modelling and the inferences are not based on sampling distributions</font>
    
<font size = "1"> For a fuller description see the Practicing Bayesian statistics markdown file within the Github repository.</font>

# Steps of Bayesian data analysis

<font size = "3"> Kruschke (2015) offers a step by step formulation for how to conduct a Bayesian analysis:

1. Identify the relevant data for question under investigation.

2. Define the descriptive (mathematical) model for the data.

3. Specify the Priors for the model. In the case of scientific research publication is the goal, as such the priors must be accepted by a skeptical audience. Much of this can be achieved using prior predcitve checks to ascertain if the priors are reasonable.

4. Using Bayes rule estimate the posterior for the parameters of the model using the likelihood and priors. Then interpret the parameter posteriors.

5. Conduct model checks. i.e. Posterior predictive checks.</font> 

<font size = "1">This notebook will follow this approach generally.</font>

## Study Description

The background for the data analysed below, is that experts play an ever increasing role in a ever increasing complex society with focused specialisation of said experts. A porposed phenomenena is that experts may be over convinced of their capabilities and know more than they actually do and are vulnerable to illusionary thoughts. All resulting in overclaiming(e.g. claiming impossible things).

Atir, Rosenzweig and Dunning (2015) study took 202 online participants and exposed them to a overclaiming task. Participants completed a overclaiming task where participants responded with how much they each new about 15 finance terms. 3 of these terms did not exist. This is where any overclaiming was measured. The second task was self- percieved knowledge (personnel finance) questionaire. These two tasks were counterbalanced to control for order effects.

For the analysis we are going to analyse the associations between two predictor varaiables on the proportion of overclaiming. The first of the 2 is the self-percieved knowledge score and the second is the FNIRA score.

The model below is an example of a model that uses data informed default priors that are weakly infomaative and provide regularised inferences (as sepcified by https://cran.r-project.org/web/packages/rstanarm/vignettes/priors.html#default-priors-and-scale-adjustments). The choice to use these defualts here is that it demonstrates this analytical option, as well as, demonstrating my own ignorance for how to set priors for this particular analysis problem, whilst at the same time attempting to avoid the the use of uniform flat priors.

This use of defaults makes prior predictive checks mute as priors here are not aiming at placing resonable prior probabiltiy to the model parameters to the DGP outcome space, but to simple reguralise the inference. Finally it is important to point out despite the use of priors here the nature of them being defaults informed by data means the analysis is no longer truely Bayesian (to fullest extent) and falls more towards an empirically Bayesian analysis as the priors are informed by the data.

# Step 2 - Define the descriptive statistical model  \begin{align*}
y_i &\sim Normal(\mu_i, \sigma) 
\\ \mu_i &= \beta_0 + \beta_1(x_1-\bar{x}_{1}) + \beta_2(x_2-\bar{x}_2)
\\ \beta_0 &\sim normal(m_y, 2.5 \cdot s_y)
\\ \beta_1 &\sim normal(0, 2.5\cdot s_y/s_x{_1})
\\ \beta_2 &\sim normal(0, 2.5\cdot s_y/s_x{_2})
\\ \sigma &\sim exponential(1/s_y) 
\end{align*} 

As the model above shows the priors are informed by the data, because the sample means and standard deviations are used to set the hyperparameters of the priors. What is also showed is that these defaults are based on centering the predictor variables. Therefore, the estimated parameter for the intercept is to be interpreted as expectation values of the dependent variable when all of the predictors are equal to the respective expected values. Either way the $beta$ slopes do not change due to mean centering and are primary interest.

```python
# Stan model for conducting multiple regression
Multiple_Regression = """
data {
  int<lower=0> N;   // Sample size
  int<lower=0> K;   // Number of predictor variables
  matrix[N, K] x;   // Predictor variables (IV) matrix
  vector[N] y;      // Vector of Dependent variable (DV) values
  
  // prior settings for beta coeffiecents
  vector[K] mu_PS_beta;
  vector[K] sd_PS_beta;
  
  // prior settings for sigma coeffiecents
  real lamda_PS_sigma;
}

transformed data{
// Creating a new design matrix with the Independent variable of Inherence bias
// centered. 
matrix[N, K] x_transformed = x;

// Center the predictor variable
for (i in 2:K){
    x_transformed[,i] = x[,i] - mean(x[,i]);
    }
}

parameters {
    
  vector[K] beta;       // Coefficients for predictors
  real<lower=0> sigma;  // Standard deviation
}

model {
// Default priors (vectorised)
beta ~ normal(mu_PS_beta,sd_PS_beta);
sigma ~ exponential(lamda_PS_sigma);
// likelihood
y ~ normal(x_transformed * beta, sigma);  
}

generated quantities {
// Generate simulated datasets based on the model to conduct posterior predictive checks
real yrep[N];
yrep = normal_rng(x_transformed * beta, sigma);

}
"""
```

```python
# Generate design matrix for the regression analysis for use in the stan model above
x = np.asarray(pt.dmatrix(' ~ self_perceived_knowledge + accuracy', data = df))
```

```python
# Generating a python dictionary for passing to the stan model specifiesd above.
data = {
    'N': len(df),
    'K': x.shape[1],
    'x': x,
    'y': df["overclaiming_proportion"].values,
    
    # Set prior values outside of stan based on rstanarm implementation.
    # Priors for the beta parameter prior means
    'mu_PS_beta': [np.mean(df["overclaiming_proportion"]), 0., 0.],
    
    # Priors for the beta parameter prior standard deviations
    'sd_PS_beta': [2.5 * np.std(df["overclaiming_proportion"]), 
                   2.5 * np.std(df["overclaiming_proportion"])/np.std(x[0]), 
                   2.5 * np.std(df["overclaiming_proportion"])/np.std(x[1])],
    
    # Prior value for the sigma parameter
    'lamda_PS_sigma': 1/np.std(df["overclaiming_proportion"])   
}
```

```python
# Code below compiles the Stan model specified above. Warnings are bug with stan
# prior predictive checks reveal prior on the parameters are recognised when run.
sm = ps.build(Multiple_Regression, data = data)
```

```python
# Fiting model to the complied stan model above with 4 chains 
fit = sm.sample(num_chains = 4, num_samples = 1000)
```

```python
fit_df = fit.to_frame()
az.summary(fit, var_names = ['beta', 'sigma'])
```

# Maximum aposterior estimate (MAP) for the fitted regression model 

```python
# Plot MAP line for accuracy covariate 
x_vals1 = np.linspace(1,7,100)
xbar1 = np.mean(df['self_perceived_knowledge'])
y_vals1 = np.mean(fit_df['beta.1']) + np.mean(fit_df['beta.2']) * (x_vals1 - xbar1)
sns.scatterplot(data = df, x = 'self_perceived_knowledge', y = 'overclaiming_proportion', alpha = .9);
sns.despine(top = True, right = True,);
plt.plot(x_vals1, y_vals1, '-');
plt.xlabel('Self percieved knowledge');
plt.ylabel('Overclaiming proportion');
```

```python
# Plot MAP line for accuracy covariate 
x_vals2 = np.linspace(-.2, 1, 100)
xbar2 = np.mean(df['accuracy'])
y_vals2 = np.mean(fit_df['beta.1']) + np.mean(fit_df['beta.3']) * (x_vals2 - xbar2)
sns.scatterplot(data = df, x = 'accuracy', y = 'overclaiming_proportion', alpha = .9);
sns.despine(top = True, right = True,);
plt.plot(x_vals2, y_vals2, '-');
plt.xlabel('Accuracy score');
plt.ylabel('Overclaiming proportion');
```

# Uncertainty in regression estimates

```python
for i in range(0, 2000):
    # Mean centred
    y_vals1 = np.mean(fit_df['beta.1'][i]) + np.mean(fit_df['beta.2'][i]) * (x_vals1 - xbar1)
    plt.plot(x_vals1, y_vals1, '-y', alpha=0.01)

# Plot data points on the scatterplot.
sns.scatterplot(data = df, x = 'self_perceived_knowledge', y = 'overclaiming_proportion', alpha = .9)
sns.despine(top = True, right = True,);
plt.xlabel('Self percieved knowledge');
plt.ylabel('Overclaiming proportion');
```

```python
# Plot regression lines.
for i in range(0, 2000):
    # Mean centred
    y_vals2 = np.mean(fit_df['beta.1'][i]) + np.mean(fit_df['beta.3'][i]) * (x_vals2 - xbar2)
    plt.plot(x_vals2, y_vals2, '-y', alpha=0.01)

# Plot data points on the scatterplot.
sns.scatterplot(data = df, x = 'accuracy', y = 'overclaiming_proportion', alpha = .9);
sns.despine(top = True, right = True,);
plt.xlabel('Accuracy score');
plt.ylabel('Overclaiming proportion');
```

# Step 5 - Posterior predictive checks

```python
#Convert pystan fit object to IO for Arviz functions.
idata = az.from_pystan(posterior=fit,posterior_model = sm, posterior_predictive=['yrep'],observed_data= 'y')
```

```python
# Plot posterior simulated data sets for posterior predictive check
az.plot_ppc(idata, data_pairs = {"y" : "yrep"}, num_pp_samples= 100);
```

As the posterior predictive checks show above, there is a clear deficiency in using a normal likelihood multiple regression model to analyze the data here. The model poorly captures the data, mainly because the dependent variable is a proportion score. If enough data had been available, the normal model might have produced a reasonable approximation. An alternative to multiple normal regression would be to use a beta regression (Liu & Kong, 2015). This type of model is demonstrated in its own notebook within the repository.

# Reporting the results of the Bayesian mulitple egression

<font size="3">As Kruschke correctly points out, there is no standard formula or presentation method for the results of a Bayesian data analysis in a journal article, like the APA guide for reporting a frequentist analysis. It is unlikely there ever will be because, as McElreath (2020) explains, Bayesian data analysis is more of an engineering approach to the problem, and the resulting model that is fit will likely be analysis-specific. In addition, as Gabry et al. (2019) argue, visualizations may be even more key, so all the visualizations above would have to be included with any write-up. Anyway, the write-up below generally follows the advice of Kruschke (2015) chapter 25. In any application, though, it comes down to the problem to be described and the audience that needs to be convinced.</font>

<h2>Write up</h2><br/>

Ignoring the poor fit of the model here, the following right up is still a refrence example for reporting a Bayesian multiple regression model with a normal likelihood function.

The overclaiming proportion scores were analyzed here using a multiple regression model with a normal likelihood. The model had two predictor variables (self-perceived knowledge and accuracy). Overall, this resulted in a model that contained 4 parameters. These parameters were the intercept $\beta_0$ and two slope parameters $\beta_1$ and $\beta_2$, and the standard deviation parameter $\sigma$. The priors for the model were selected based on the defaults provided by Goodrich, Gabry, Ali, and Brilleman (2018). See the statistical formula above for further specifics. The chosen priors are data-informed and designed to provide regularization for inferences. In addition, the defaults have been selected due to a lack of problem-specific knowledge that would enable the elicitation of problem-specific priors by the analysts.

Four MCMC chains were run, each taking 2000 samples from the posterior. 1000 of the samples from each chain were used as warm-up and discarded from the final analysis (i.e., not used in the inferences). Convergence of the MCMC chains was examined through the use of autocorrelation and trace plots. Visual inspection suggested no issues in terms of autocorrelation or chain mixing, and the generally accepted metric of R-hat < 1.05 and an effective sample size > 400 were achieved for all estimated parameters (see Vehtari et al. for further discussion).

Analysis of the posterior showed for the parameters of interest the slopes $\beta_1$ and $\beta_2$ that $\beta_1$ (self-perceived knowledge = $\mu$ = 0.01, with a 94% CrI [0.06, 0.14]) and $\beta_2$ (accuracy) = $\mu$ = -0.67, with a 94% CrI [-0.858, -0.472]. These results suggest, based on general indices of effect under the Bayesian framework (Kruschke, 2018; Makowski, Ben-Shachar, Chen, & Ldecke, 2019), that because the credible interval (94% default of arviz package) does not cross 0, both predictors significantly predict overclaiming proportion, albeit in opposite directions. Such that increases in perceived self-knowledge were associated with increases in the proportion of overclaiming, whereas accuracy was negatively associated with overclaiming proportion.
